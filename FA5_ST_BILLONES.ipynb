```python
#2

#8.34

arrpop = [3, 7, 11, 15]

arrpop_sum=0
for i in range(len(arrpop)):
    arrpop_sum += (arrpop[i])

arrpop_mean= arrpop_sum/len(arrpop)
print(f"(a) mean of the population: {arrpop_mean}")

std_sum_arrpop=0
for i in range(len(arrpop)):
    std_sum_arrpop += ((arrpop[i])-arrpop_mean)*((arrpop[i])-arrpop_mean)

std_arrpop= std_sum_arrpop/len(arrpop)
print(f"(b) standard deviation of the population: {std_arrpop}")

import itertools
combinations = list(itertools.product(arrpop, repeat=2))

# Calculate the means for each combination
cormean = [sum(combination) / 2 for combination in combinations]

# Print the corresponding means
print("(c) Corresponding means:", cormean)

sum_mean_arrpop=0
for i in range(len(cormean)):
    sum_mean_arrpop += ((cormean[i]))


mean_value= sum_mean_arrpop / len(combinations)
print("The mean of Sampling Mean is:", mean_value)

newstd_sum_arrpop=0
for i in range(len(combinations)):
    newstd_sum_arrpop += ((cormean[i])-arrpop_mean)*((cormean[i])-arrpop_mean)

newstd_arrpop= newstd_sum_arrpop/len(combinations)  
print("(d) standard deviation of the sampling distribution of mean is:", newstd_arrpop)
```

```python
#2

#8.21

import scipy.stats as stats

# Probability of a boy (p)
p = 0.5

# Total number of children (n)
n = 200

# (a) 
k_a = int(0.40 * n)
prob_less_than_40_percent = sum(stats.binom.pmf(range(k_a), n, p))
print(f"(a) Probability that less than 40% will be boys: {prob_less_than_40_percent:.4f}")

# (b) 
k_b_lower = int(0.43 * n)
k_b_upper = int(0.57 * n)
prob_between_43_and_57_percent = sum(stats.binom.pmf(range(k_b_lower, k_b_upper + 1), n, p))
print(f"(b) Probability that between 43% and 57% will be girls: {prob_between_43_and_57_percent:.4f}")

# (c) 
k_c = int(0.54 * n)
prob_more_than_54_percent = 1 - sum(stats.binom.pmf(range(k_c + 1), n, p))
print(f"(c) Probability that more than 54% will be boys: {prob_more_than_54_percent:.4f}")
```

```python
#3 

#8.49

x_values = [6, 9, 12, 15, 18]
p_values = [0.1, 0.2, 0.4, 0.2, 0.1]

# (μ)
mean = sum(x * p for x, p in zip(x_values, p_values))
print(f"Mean (μ): {mean}")

# (σ²)
variance = sum((x - mean) ** 2 * p for x, p in zip(x_values, p_values))
print(f"Variance (σ²): {variance}\n")


import itertools

# sample_combinations
sample_combinations = list(itertools.product(x_values, repeat=2))

# Means and probabilities for each sample
sample_means = [(x1 + x2) / 2 for x1, x2 in sample_combinations]
sample_probabilities = [p1 * p2 for p1, p2 in itertools.product(p_values, repeat=2)]

# 25 samples, their means, and probabilities
for i in range(25):
    print(f"Sample {i+1}: {sample_combinations[i]}, Mean: {sample_means[i]}, Probability: {sample_probabilities[i]}")
```
